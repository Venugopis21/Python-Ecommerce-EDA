## Q 5. Find the net revenue generated by the E-Commerce company over all orders placed. (6 marks)

### Ans 5

```python
# all brands have sold less orders.
ECom_Data_revenue=ECom_Data.groupby(['Brand']).aggregate({'Order_Date':'count','discounted_price':'sum'})
ECom_Data_revenue
```

```python
# No brand has sold above 600 orders
filter1 = ECom_Data_revenue["Order_Date"]>600
filter1
```

```python
# Allure Auto falls in the 2nd category where 15% of orders are charged
filter2 = ECom_Data_revenue[(ECom_Data_revenue.Order_Date>350) & (ECom_Data_revenue.Order_Date<=600)]
filter2
```

```python
filter2['discounted_price'] = filter2['discounted_price'] * 0.15
filter2
```

```python
revenue1 = filter2['discounted_price'].sum()
print(revenue1)
```

```python
# These following brands contribute to the 3rd category where 10% of orders are charged.
# Apart from these rest all brands are charged 5% of totol orders.
filter3 = ECom_Data_revenue[(ECom_Data_revenue.Order_Date>100) & (ECom_Data_revenue.Order_Date<=350)]
filter3
```

```python
filter3['discounted_price'] = filter3['discounted_price'] * 0.10
filter3
```

```python
revenue2 = filter3['discounted_price'].sum()
print(revenue2)
```

```python
filter4 = ECom_Data_revenue[ECom_Data_revenue.Order_Date<100]
filter4
```

```python
filter4['discounted_price'] = filter4['discounted_price'] * 0.05
filter4
```

```python
revenue3 = filter4['discounted_price'].sum()
print(revenue3)
```

```python
# 615446.95 is the revenue made by the company.
Total_revenue = revenue1 + revenue2 + revenue3
print(Total_revenue)
```

```python
Total_revenue_v2 = pd.concat([filter2, filter3,filter4])
Total_revenue_v2
```

```python
Total_revenue_v3 = Total_revenue_v2['discounted_price'].sum()
print(Total_revenue_v3)
```

```python
filter2_brandrevenue['Total_brand_revenue'] = filter2_brandrevenue['discounted_price'] * 0.85
filter2_brandrevenue
```

```python
# These following brands contribute to the 3rd category where 10% of orders are charged.
# Apart from these rest all brands are charged 5% of totol orders.
filter3_brandrevenue = ECom_Data_revenue[(ECom_Data_revenue.Order_Date>100) & (ECom_Data_revenue.Order_Date<=350)]
filter3_brandrevenue
```

```python
filter3_brandrevenue['Total_brand_revenue'] = filter3_brandrevenue['discounted_price'] * 0.90
filter3_brandrevenue
```

```python
filter4_brandrevenue = ECom_Data_revenue[ECom_Data_revenue.Order_Date<100]
filter4_brandrevenue
```

```python
filter4_brandrevenue['Total_brand_revenue'] = filter4_brandrevenue['discounted_price'] * 0.95
filter4_brandrevenue
```

```python
# Brand Revenue for each brand has been calculate here
Brand_Revenue = pd.concat([filter2_brandrevenue, filter3_brandrevenue,filter4_brandrevenue])
Brand_Revenue
```
